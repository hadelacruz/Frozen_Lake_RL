## Task 2.2: Algoritmo Q-Learning

### Hiperparámetros Utilizados

| Parámetro | Símbolo | Valor |
|-----------|---------|-------|
| Learning Rate | α | 0.1 |
| Discount Factor | γ | 0.99 |
| Epsilon Inicial | ε₀ | 1.0 |
| Epsilon Mínimo | ε_min | 0.01 |
| Epsilon Decay | decay | 0.995 |
| Episodios | N | 10,000 |

### Fórmula de Actualización Q-Learning

$$Q(s,a) \leftarrow Q(s,a) + \alpha[R + \gamma \max_{a'} Q(s',a') - Q(s,a)]$$

Donde:
- $s$ = estado actual
- $a$ = acción tomada
- $R$ = recompensa recibida
- $s'$ = siguiente estado
- $\alpha$ = learning rate
- $\gamma$ = discount factor
